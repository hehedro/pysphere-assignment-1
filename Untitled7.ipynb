```python
# Welcome Sequence
print("Welcome to the Voltage Calculator")
#Define the formula for calculating Voltage
def V(I,R):
  return I * R
# Collecting the value of current in Amperes
I = input("What is the current in Amperes? ")
#saving the value of current as a float
I = float(I)
# Checking the unit of Resistance
is_ohms = input("Is the resistance value in Ohms? ").lower()
# finding out the unit resistance is in if not Ohms
if is_ohms == "no":
  new_answer = input("Is it in (a) kilohms or (b) megohms? ").lower()
  if new_answer == "a":
# Correcting the value of Resistance not previously in ohms to ohms
    Kilohms = input("What is the resistance in Kilohms ")
    R = float(Kilohms) * 1000
  elif new_answer == "b":
    Megohms = input("What is the resistance in Megohms ")
    R = float(Megohms) * 1000000
  else:
    print("Invalid input")
elif is_ohms == "yes":
  R = input("What is the resistance in Ohms ")
  R = float(R)
else:
  print("Invalid input")
print(f"The voltage(V) is {V(I,R)}")
```
